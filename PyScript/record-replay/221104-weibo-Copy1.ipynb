```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
files = ["raw", "10", "20", "30", "40", "50"]
```

```python
all_time_pd = [pd.read_csv("./plTime/weibo.com.%s" % x, sep = '\t', header = None) for x in files]
```

```python
for idx, x in enumerate(all_time_pd):
    sns.ecdfplot(x[x[0] == "PLT"][1], label = files[idx])
plt.grid()
plt.legend()
```
